```python
img = cv2.imread("Capture.PNG",0)
plt.imshow(img, cmap="gray")
```

```python
lim,img1 = cv2.threshold(img,100,255,cv2.THRESH_BINARY)
```

```python
plt.imshow(img1,cmap="gray")
```

```python
lim,img2 = cv2.threshold(img,100,255,cv2.THRESH_BINARY_INV)
plt.imshow(img2,cmap="gray")
```

```python
lim,img3 = cv2.threshold(img,100,255,cv2.THRESH_TRUNC)
plt.imshow(img3,cmap="gray")
```

```python
lim,img4 = cv2.threshold(img,100,255,cv2.THRESH_TOZERO)
plt.imshow(img4,cmap="gray")
```

```python
lim,img5 = cv2.threshold(img,100,255,cv2.THRESH_TOZERO_INV)
plt.imshow(img5,cmap="gray")
```
